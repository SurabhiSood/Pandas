```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#student_name	gender	grade	school_name	reading_score	math_score	School ID	type	size	budget

df=school_data_complete.copy()

Total_Schools=df["school_name"].nunique()
Total_Students=df["Student ID"].nunique()

Budget=df.drop_duplicates("budget")
Total_Budget=Budget["budget"].sum()

Average_Math_Score=df["math_score"].mean()
Average_Readind_Score=df["reading_score"].mean()
Overall_Passing_Rate=(Average_Math_Score+Average_Readind_Score)/2
print(Overall_Passing_Rate)

#Calculate the percentage of students with a passing math score (70 or greater)
math_pass=df['math_score'].loc[df['math_score']>=70].count()
math_percent=(math_pass/Total_Students)*100
print(math_percent)

#Calculate the percentage of students with a passing reading score (70 or greater)
read_pass=df['reading_score'].loc[df['reading_score']>=70].count()
read_percent=(read_pass/Total_Students)*100
print(read_percent)

New_df=pd.DataFrame({"Total Schools":[Total_Schools],
                    "Total Students":[Total_Students],
                    "Total Budget":[Total_Budget],
                    "Average Math Score":[Average_Math_Score],
                    "Average Reading Score":[Average_Readind_Score],
                    "% Passing Math":[math_percent],
                    "% Passing Read":[read_percent],
                    "% Overall Passing Rate":[Overall_Passing_Rate]})

New_df

#data["Age"]= data["Age"].replace(25.0, "Twenty five") 
#weeks = training_df["Membership (Days)"]/7
#training_df["Membership (Weeks)"] = weeks
#df["math_score"]=np.where(df['math_score'].between(0,69), "Fail", df['math_score']



```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
#School Type	Total Students	Total School Budget	Per Student Budget	
#Average Math Score	Average Reading Score	% Passing Math	% Passing Reading	% Overall Passing Rate

#grouped_passing_math = school_data_complete[school_data_complete['math_score']>=70].groupby(['school_name']).size()

Performance_df=school_data_complete.copy()
School_Type=Performance_df.groupby("school_name")['type'].first()
Total_students=Performance_df.groupby("school_name").size()
Total_School_budget=Performance_df.groupby("school_name")['budget'].first()
Per_student_budget=Total_School_budget/Total_students
Math_Score=Performance_df.groupby("school_name")['math_score'].sum()
Average_Math_Score=Math_Score/Total_students
Reading_Score=Performance_df.groupby("school_name")['reading_score'].sum()
Average_Reading_Score=Reading_Score/Total_students
Passing_Math=school_data_complete[school_data_complete['math_score']>=70].groupby(['school_name']).size()
Percent_Math=Passing_Math/Total_students*100
Passing_Read=school_data_complete[school_data_complete['reading_score']>=70].groupby(['school_name']).size()
Percent_Read=Passing_Read/Total_students*100
Passing_Rate=(Percent_Math+Percent_Read)/2

Performance_summary=pd.DataFrame({"School type":School_Type,
                                  "Total Students":Total_students,
                                  "Total School Budget":Total_School_budget,
                                  "Per Student Budget":Per_student_budget,
                                  "Average Math Score":Average_Math_Score,
                                  "Average Reading Score":Average_Reading_Score,
                                  "% Passing Math":Percent_Math,
                                  "% Passing Reading":Percent_Read,
                                  "% Overall Passing Rate":Passing_Rate})

calculation_summary=Performance_summary.copy()

Performance_summary=Performance_summary.sort_values(['% Overall Passing Rate'],ascending=False)
Performance_summary["% Passing Math"]=Performance_summary["% Passing Math"].map("{:.2f}%".format)
Performance_summary["% Passing Reading"]=Performance_summary["% Passing Reading"].map("{:.2f}%".format)
Performance_summary["% Overall Passing Rate"]=Performance_summary["% Overall Passing Rate"].map("{:.2f}%".format)
Performance_summary.head()
```

## Bottom Performing Schools (By Passing Rate)

* Sort and display the five worst-performing schools

```python
Performance_summary=Performance_summary.sort_values(['% Overall Passing Rate'],ascending=True)
Performance_summary=Performance_summary.reset_index()
Performance_summary.head()
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
math_score_per_grade=school_data_complete.groupby(['grade','school_name'])['math_score'].sum()
students_per_grade=school_data_complete.groupby(['grade','school_name'])['Student ID'].count()
average_per_grade=math_score_per_grade/students_per_grade
                                     
grade_df=pd.DataFrame({"Average Math Score pg":average_per_grade,})
grade_df=grade_df.reset_index()
grade_df=grade_df.pivot(index='school_name',columns='grade',values='Average Math Score pg')
grade_df
                                    
```

## Reading Score by Grade 

* Perform the same operations as above for reading scores

```python
reading_score_per_grade=school_data_complete.groupby(['grade','school_name'])['reading_score'].sum()
students_per_grades=school_data_complete.groupby(['grade','school_name'])['Student ID'].count()
average_per_grades=reading_score_per_grade/students_per_grade
                                     
Read_grade_df=pd.DataFrame({"Average Reading Score pg":average_per_grades,})
Read_grade_df=Read_grade_df.reset_index()
Read_grade_df=Read_grade_df.pivot(index='school_name',columns='grade',values='Average Reading Score pg')
Read_grade_df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]

New_score_df=calculation_summary.copy()
New_score_df['Average Spending Range (ps)']=pd.cut(New_score_df['Per Student Budget'],bins, labels=group_names, include_lowest=True)

Score_math=New_score_df.groupby('Average Spending Range (ps)')['Average Math Score'].mean()
Score_read=New_score_df.groupby('Average Spending Range (ps)')['Average Reading Score'].mean()

p_math=New_score_df.groupby('Average Spending Range (ps)')['% Passing Math'].mean()
p_read=New_score_df.groupby('Average Spending Range (ps)')['% Passing Reading'].mean()
p_overall=New_score_df.groupby('Average Spending Range (ps)')['% Overall Passing Rate'].mean()

Spending_df=pd.DataFrame({"Average Math Score":Score_math,"Average Reading Score":Score_read,
                          "% Passing Math":p_math,
                          "% Passing Reading":p_read,
                          "Overall Passing Rate":p_overall
                         })
Spending_df.head()
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
# Sample bins. Feel free to create your own bins.
bins = [0, 1000, 2000, 5000]
group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
New_school_size_df=calculation_summary.copy()
New_school_size_df['Scores by School Size']=pd.cut(New_school_size_df['Total Students'],bins, labels=group_names, include_lowest=True)

math_ss=New_school_size_df.groupby('Scores by School Size')['Average Math Score'].mean()
read_ss=New_school_size_df.groupby('Scores by School Size')['Average Reading Score'].mean()

mathp_ss=New_school_size_df.groupby('Scores by School Size')['% Passing Math'].mean()
readp_ss=New_school_size_df.groupby('Scores by School Size')['% Passing Reading'].mean()
overallp_ss=New_school_size_df.groupby('Scores by School Size')['% Overall Passing Rate'].mean()


school_size_df=pd.DataFrame({"Average Math Score":math_ss,"Average Reading Score":read_ss,
                            "% Passing Math":mathp_ss,
                            "% Passing Reading":readp_ss,
                            "Overall Passing Rate":overallp_ss
                            })
school_size_df
```

## Scores by School Type

* Perform the same operations as above, based on school type.

```python
Type_df=calculation_summary.copy()

math_st=Type_df.groupby('School type')['Average Math Score'].mean()
read_st=Type_df.groupby('School type')['Average Reading Score'].mean()
pm_st=Type_df.groupby('School type')['% Passing Math'].mean()
pr_st=Type_df.groupby('School type')['% Passing Reading'].mean()
o_st=Type_df.groupby('School type')['% Overall Passing Rate'].mean()

Score_type_df=pd.DataFrame({"Average Math Score":math_st,
             "Average Reading Score":read_st,
             "% Passing Math":pm_st,
             "% Passing Reading":pr_st,
             "% Overall Passing Rate":o_st})
Score_type_df
```
